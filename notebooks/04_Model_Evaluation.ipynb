```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

results = pd.read_csv("../results/model_metrics.csv")

baseline_pred = pd.read_csv("../results/baseline_predictions.csv")
lr_pred = pd.read_csv("../results/linear_regression_predictions.csv")
rf_pred = pd.read_csv("../results/random_forest_predictions.csv")
xgb_pred = pd.read_csv("../results/xgboost_predictions.csv")
```

```python
#Model Comparison

#MAE
plt.figure(figsize=(10,5))

plt.bar(results["Model"], results["MAE"])

plt.title("Model Comparison - MAE")
plt.xticks(rotation=45)
plt.ylabel("MAE")

plt.show()

#RMSE
plt.figure(figsize=(10,5))

plt.bar(results["Model"], results["RMSE"])

plt.title("Model Comparison - RMSE")
plt.xticks(rotation=45)
plt.ylabel("RMSE")

plt.show()

#R2
plt.figure(figsize=(10,5))

plt.bar(results["Model"], results["R²"])

plt.title("Model Comparison - R² Score")
plt.xticks(rotation=45)
plt.ylabel("R²")

plt.show()
```

```python
#Actual VS Predicted
plt.figure(figsize=(14,6))

plt.plot(xgb_pred["Actual"].values[:200], label="Actual")
plt.plot(xgb_pred["Predicted"].values[:200], label="XGBoost Predicted")

plt.title("XGBoost: Actual vs Predicted Sales (First 200 Points)")
plt.legend()

plt.show()
```

```python
#Fewer points
plt.figure(figsize=(12,6))

plt.plot(xgb_pred["Actual"].values[:50], label="Actual")
plt.plot(xgb_pred["Predicted"].values[:50], label="XGBoost Predicted")

plt.title("XGBoost: Actual vs Predicted Sales")
plt.legend()

plt.show()
```

```python
#All Models
plt.figure(figsize=(14,6))

plt.plot(xgb_pred["Actual"].values[:200], label="Actual", linewidth=2)
plt.plot(lr_pred["Predicted"].values[:200], label="Linear Regression")
plt.plot(rf_pred["Predicted"].values[:200], label="Random Forest")
plt.plot(xgb_pred["Predicted"].values[:200], label="XGBoost")

plt.title("Model Comparison (First 200 Points)")
plt.legend()

plt.show()
```

```python
## Key Insights

# Feature engineering (lag + rolling statistics) had the largest impact on model performance.
# All models performed significantly better after adding temporal features.
# XGBoost achieved the best performance (R² ≈ 0.99), but improvements over Random Forest were marginal.
# This suggests that the dataset is highly autocorrelated and historical sales patterns are strong predictors of future demand.

## Final Model Selection

#XGBoost was selected as the final model due to its lowest MAE and RMSE and highest R² score.
```
